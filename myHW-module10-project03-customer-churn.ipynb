```python
#b.extract 15th column and store in 'customer_15':

customer_15 = df.iloc[[14],:]

customer_15
```

```python
#c.extract all male senior sitizens whose payment method is Electronic check and store in 'senior_male_electronic'

senior_male_electronic = df[(df['SeniorCitizen']==1) & 
                              (df['gender'] == 'Male') & 
                              (df['PaymentMethod']=='Electronic check')]

senior_male_electronic
```

```python
#d.extract customers with tenure>70 months or their monthly charges are >100 $:
#Store result in 'customer_total_tenure'

customer_total_tenure = df[(df['tenure'] > 70) & 
                              (df['MonthlyCharges'] > 100)]
customer_total_tenure
```

```python
#e.extract customers whose Contract is 2 years, payment method is Mailed check, and value of Churn is Yes.
#Store in 'two-male-yes'

two_male_yes = df[(df['Contract'] == 'Two year') & (df['PaymentMethod'] == 'Electronic check') & 
                              (df['Churn'] =='Yes')]

two_male_yes
```

```python
#f.extract 333 random records from customer_churn dataframe and store the result in 'customer_333':

customer_333 = df.sample(n=333)

customer_333
```

```python
#B. Daza Viz
#a.buil dbar plot for InternerService column
#xlabel "categores of internwet service"
#ylabel 'count of categories'
#title of plot 'Distribution of Internet service'
#color of bars 'orange'

x = df['InternetService'].value_counts().keys()
y = df['InternetService'].value_counts()
plt.bar(x,y,color='orange')
plt.xlabel('Categories of Internet Service')
plt.ylabel('Count of Categories')  
plt.title('Distribution of Internet Service')
```

```python
"""b.histogram for 'tenure' column
number of bins 30, color of bins 'green', title "Distribution of Tenure"

"""
plt.hist(df['tenure'],color='green', bins=30)
plt.title('Distribution of tenure')
```

```python
'''
c. scatter plot between 'monthly charges' (y) and tenure (x). 
points of clolor 'brown', xlabel "Tenure of customer", ylabel "Monthly charges of customer".
Title "Tenure vs Monthly charges"
'''

plt.scatter(x = df['tenure'].head(30), y = df['MonthlyCharges'].head(30), color ='Brown')
plt.xlabel('Tenure of customer"')
plt.ylabel('Monthly Charges of customer')
plt.title('Tenure vs Monthly Charges')
```

```python
#d.box plot 'tenure' (y) and 'contract'(x)
df.boxplot(column='tenure', by=['Contract'])
plt.show()
```

```python

#C. linear regression where dependant var is "Monthly charges ", independant var is Tenure. find rmse:

from sklearn.model_selection import train_test_split
x=pd.DataFrame(df['tenure'])
y=df['MonthlyCharges']

x_train, x_test, y_train, y_test=train_test_split(x, y, test_size=0.3, random_state= 0)

#x
#y
#x_train
#x_test
#y_train
#y_test
```

```python
from sklearn.linear_model import LinearRegression
LR= LinearRegression()
LR.fit(x_train, y_train)

y_predict= LR.predict(x_test)
y_predict
```

```python
y_test
```

```python
from sklearn.metrics import mean_squared_error
mse= mean_squared_error(y_predict, y_test)
rmse=np.sqrt(mse)
rmse
```

```python
#D. a. Logistic regression.  dependant var is 'Churn', indep 'Monthly charges'. 65/35 train/test
#get confusion matriz and accuracy score:

x1=pd.DataFrame(df['MonthlyCharges'])
y1=df['Churn']

x1_train, x1_test, y1_train, y1_test=train_test_split(x1, y1, test_size=0.35, random_state= 0)

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score
LoR= LogisticRegression()
LoR.fit(x1_train, y1_train)

y1_predict=LoR.predict(x1_test)
y1_predict
```

```python
confusion_matrix(y1_predict, y1_test), accuracy_score(y1_predict, y1_test)
# TP FP
#FN TN
```

```python
#.D. b.Multiple Logistic regression.  dependant var is 'Churn', indep 'tenure' and Monthly charges'.80/20.
#get confusion matriz and accuracy score:

x2=pd.DataFrame(df.loc[:,['MonthlyCharges','tenure']])
y2=df['Churn']

x2_train, x2_test, y2_train, y2_test=train_test_split(x2, y2, test_size=0.2, random_state= 0)
x2_train
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score
LoR= LogisticRegression()
LoR.fit(x2_train, y2_train)

y2_predict=LoR.predict(x2_test)
y2_predict
```

```python
confusion_matrix(y2_predict, y2_test), accuracy_score(y2_predict, y2_test)
```

```python
# E. Decision tree.dependant var is 'Churn', indep 'tenure' 80/20, confusion matriz and accuracy score:

x3=pd.DataFrame(df['tenure'])
y3=df['Churn']

x3_train, x3_test, y3_train, y3_test=train_test_split(x3, y3, test_size=0.2, random_state= 0)
```

```python
from sklearn.tree import DecisionTreeClassifier
DecisionTree= DecisionTreeClassifier()
DecisionTree.fit(x3_train, y3_train)

y3_predict=DecisionTree.predict(x3_test)
y3_predict
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
confusion_matrix(y3_predict, y3_test), accuracy_score(y3_predict, y3_test)
```

```python
#F. Random Forest. dependant var is 'Churn', indep 'tenure' and Monthly charges. 70/30.
#confusion matriz and accuracy score:

x4=pd.DataFrame(df.loc[:,['MonthlyCharges','tenure']])
y4=df['Churn']

x4_train, x4_test, y4_train, y4_test=train_test_split(x4, y4, test_size=0.3, random_state= 0)
```

```python
from sklearn.ensemble import RandomForestClassifier
RFC=RandomForestClassifier(n_estimators=100)
RFC.fit(x4_train, y4_train)


y4_predict=RFC.predict(x4_test)
y4_predict
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
confusion_matrix(y4_predict, y4_test), accuracy_score(y4_predict, y4_test)
```
