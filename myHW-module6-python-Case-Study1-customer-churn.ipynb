```python
#task 4: box plot between "Monthly Charges" and "Payment Method":

# Create box plot in python with fills and labels
#Electronic check             2365
#Mailed check                 1612
#Bank transfer (automatic)    1544
#Credit card (automatic)      1522

value1 = a1['MonthlyCharges']   #Electronic check             2365

value2 = a2['MonthlyCharges']   #Mailed check                 1612

value3 = a3['MonthlyCharges']   #Bank transfer (automatic)    1544

value4 = a4['MonthlyCharges']   #Credit card (automatic)      1522

box_plot_data = [value1, value2, value3, value4]
#'MonthlyCharges']
#'PaymentMethod']

box_plot_data = [value1,value2,value3,value4]
plt.xlabel('Payment Method of Customer')
plt.ylabel('MonthlyCharges Incured (in US$)')
plt.title('Monthly Charges vs Payment Method')

box=plt.boxplot(box_plot_data, patch_artist=True,labels=[
            'Electronic check', 'Mailed check', 'Bank transfer', 'Credit card'])
#colors = ['green', 'tan', 'olive', 'cyan']
colors = ['olive','olive','olive','olive']
for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)

plt.show()
```
