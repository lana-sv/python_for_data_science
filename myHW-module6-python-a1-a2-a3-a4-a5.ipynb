```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
#HOMEWORK MODULE 6 - MATPLOTLIB
#ASSIGNMENT1: 1) import cars.csv
# 2) line plot for columns 'model' and 'hp'
```

```python
df =pd.read_csv('cars.csv')
df.head()
```

```python
df.columns
```

```python
#map 'model' on x axis, 'hp' on y-axis

plt.figure(figsize=(10,5))
plt.plot(df['Model '],df['HP'])
plt.title('Model Names vs Horse Power for our cars')
plt.xlabel('Models of the Cars')
plt.xticks(rotation=90)
plt.ylabel('Horse Power of Cars')
plt.show()
```

```python
x=df['Model ']
y=df['HP']

plt.plot(x,y)
```

```python
#Assignment 2: bar plot grach on 'carbs' and 'gear' columns:
```

```python
plt.bar(df['Carb'],df['Gear'],color='green')
plt.title('"Carbs vs Gear"')
plt.xlabel('Number of Carburetors')
#plt.xticks(rotation=90)
plt.ylabel('Number of Forward Gears')
plt.show()
```

```python
color1=['green','red','hotpink','orange','yellow','cyan','black']
plt.bar(df['Carb'],df['Gear'],color=color1)
plt.title('"Carb vs Gear"')
plt.xlabel('Number of Carburators')
#plt.xticks(rotation=90)
plt.ylabel('Number of Gears')
plt.show()
```

```python
#assignment 3: histogram for column 'wt'
```

```python
x=df['Wt']

plt.hist(x, bins=30, histtype='bar', align='mid', color='c',label='Wt', edgecolor='black')

plt.legend()

plt.xlabel('Weight of the cars')
plt.ylabel('Count')

plt.title('Histogram for the Weight values')
plt.show()
```
