```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.cluster import AgglomerativeClustering

from sklearn.preprocessing import StandardScaler
```

```python
from sklearn import datasets
from sklearn import cluster, datasets
from sklearn.decomposition import PCA
```

```python
#homework - module 9 - assignment 3 - iris dataset - agglomerative clustering:
#load data from sklearn datasets package
#extract data from dataset
#train agglomerative clustering model
#plot dendroigram to visualize clustering linkage
```

```python
iris = datasets.load_iris() 
X=iris['data']
Y=iris.target
print(X.shape)
```

```python
#X
```

```python
#Y
```

```python
scaler = StandardScaler()
X_std = scaler.fit_transform(X)
```

```python
plt.figure(figsize=(10,8))
X = PCA(n_components=2).fit_transform(X)
plt.scatter(X[:,0],X[:,1])
```

```python
clt = AgglomerativeClustering(linkage='complete', 
                              affinity='euclidean', 
                              n_clusters=3)

# Train model
model = clt.fit(X_std)
```

```python
model.labels_
```

```python
X_iris, y_iris = datasets.load_iris(return_X_y=True)
```

```python
plt.scatter(iris.data[:,0] ,iris.data[:,1], c= model.labels_, cmap='cool')

#X[:,0],X[:,1]
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage

links = linkage(iris.data, 'single')

points = range(1, 11)

plt.figure(figsize=(8, 6))

dendrogram(links, orientation='top', labels=model.labels_, distance_sort='descending', show_leaf_counts=True)

plt.show()
```

```python
d = np.array([[5,3], [10,15], [15,12], [24,10], [30,30], [85,70], [71,80], [60,78], [70,55], [80,91]])
```

```python
points = range(1, 11)

plt.figure(figsize=(8, 6))

plt.subplots_adjust(bottom=0.2)

plt.scatter(d[:,0],d[:,1], label='True Position', color = 'r')

for point, x, y in zip(points, d[:, 0], d[:, 1]):

    plt.annotate(point, xy=(x, y), xytext=(-3, 3),textcoords='offset points', ha='right', va='bottom')

plt.show()
```

```python
#Agglomerative clustring = Hierarchical Clustering

#we ned to create a Proximity matrix

# dendrogram - how we use it?
#to deside how many clusters do we need

#initially we nad 5 clusters(branches), then 3 left, then 2 left (they are merging togerther), then 1

#or from the other side - we has all data in 1 cluster, then it split on 2, then 3, then 4, then 5


```

```python
from scipy.cluster.hierarchy import dendrogram, linkage

links = linkage(d, 'single')

points = range(1, 11)

plt.figure(figsize=(8, 6))

dendrogram(links, orientation='top', labels=points, distance_sort='descending', show_leaf_counts=True)

plt.show()
```
